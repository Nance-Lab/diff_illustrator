```python
import numpy as np
from tqdm import tqdm

def compute_msd_3d_with_progress(X, Y, Z, N):
    msd = np.zeros(N-1)  # Initialize the MSD array to store the results
    
    # Use tqdm to wrap the range function for the outer loop
    for n in tqdm(range(1, N), desc='Computing MSD'):
        sum_squared_displacement = 0
        m = N - n  # Number of displacements for this time lag
        
        for i in range(m):
            dx = X[i + n] - X[i]
            dy = Y[i + n] - Y[i]
            dz = Z[i + n] - Z[i]
            sum_squared_displacement += dx**2 + dy**2 + dz**2
        
        msd[n-1] = sum_squared_displacement / m
    
    return msd

# Example usage:
# Assuming X, Y, and Z are numpy arrays containing the X, Y, and Z positions of the particle,
# and N is the total number of frames in the trajectory.
# X = np.array([...])
# Y = np.array([...])
# Z = np.array([...])
# N = len(X)  # Assuming the lengths of X, Y, and Z are the same
# msd_values = compute_msd_3d_with_progress(X, Y, Z, N)
# print(msd_values)
```
